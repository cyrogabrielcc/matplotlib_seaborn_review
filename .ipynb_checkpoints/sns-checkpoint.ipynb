```python
import pandas as pd
import seaborn as sns
import numpy as np

# Definindo o tema
sns.set_style(style='darkgrid')
```

```python
bd = sns.load_dataset('tips')
bd.head(5)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = bd);
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = bd, hue = 'sex');
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = bd, kind='line', hue='smoker');
```

```python
sns.histplot(data=bd, x='total_bill')
```

```python
sns.histplot(data=bd, x='total_bill', hue = 'smoker')
```

```python
sns.barplot(data=bd, x='sex', y='total_bill', hue='smoker')
```

```python
sns.pairplot(bd)
```

```python
sns.pairplot(bd, hue='sex')
```

```python
sns.pairplot(bd, kind='kde')
```

```python
sns.boxplot(data=bd,x='day',y='total_bill' )
```

```python
sns.boxplot(data=bd,x='day',y='total_bill', hue='sex' )
```
