```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

y = np.random.randint(low=1000, high=3500, size=20)
```

```python
plt.plot(y)
```

```python
plt.plot(y, 
         color='#749187', #cor da linha
         marker='o', # Aresta de contato (marcador)
         ms=5, # tamanho do marcador
         mec='k',  # Cor da linha do marcador
         markerfacecolor='w', #interior do marcador (cor)
         ls='-.' # estilo da linha
        )
plt.plot(y*2, color="gray", marker = '*', ls = "-")
plt.grid(#axis='y', # direo
         color = 'red', # cor da linha 
         linestyle = '--', # Estilo da linha
         alpha = 0.3
        )
plt.xlabel("Eixo X", color = "red", size = 12)
plt.ylabel("Eixo Y", color = "red", size = 12)
plt.title("Algo Legal", size=22)
plt.show;
```

```python
x = np.arange(1,11)
y1 = np.random.randint(1,400,10)
y2 = np.random.randint(150,500,10)
y3 = np.random.randint(10,400,10)

fig, ax = plt.subplots(1,3, figsize =(15,5))
fig.suptitle('FIGURE') # subtitulo
ax[0].plot(x, y1, color="black")
ax[1].plot(x, y2, color="red")
ax[2].plot(x, y3, color="yellow")

for i in range(3):
    ax[i].set(title=f'Subplot{i+1}', xlabel = "XXXXX", ylabel = "YYYYY")
```

```python
y = np.random.uniform(0,20,100)
x = np.random.normal(10,0.5,100)
```

```python
fig = plt.figure()
ax = plt.axes()

# plotando x e y 
ax.scatter(x, y, 
           marker='o', 
           color = 'red', 
           label='data1', 
           alpha=0.5)

# plotando x e y * 0.5
ax.scatter(x*0.5, y*0.5, 
           marker='v', 
           color = 'black', 
           label='data1', 
           alpha=0.5)

ax.legend()
```

```python
import plotly.express as px

df = px.data.gapminder().query('country == "Brazil"').set_index('year')
```

```python
plt.plot(df.index, df['gdpPercap'], 
         color = 'magenta', 
         linewidth=2,
         marker='o',
         linestyle='-.',
         markersize = 5,
         markeredgecolor = 'green'
        )
plt.grid(color="grey", linestyle="--", axis='y', alpha=0.4)

plt.title('PIB per capta do Brasil')
plt.xlabel('Ano')
plt.ylabel('Valor do PIB')
```
