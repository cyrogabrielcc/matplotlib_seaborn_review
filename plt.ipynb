```python
import plotly.express as px

df = px.data.gapminder().query('country == "Brazil"').set_index('year')
```

```python
plt.plot(df.index, df['gdpPercap'], 
         color = 'magenta', 
         linewidth=2,
         marker='o',
         linestyle='-.',
         markersize = 5,
         markeredgecolor = 'green'
        )
plt.grid(color="grey", linestyle="--", axis='y', alpha=0.4)

plt.title('PIB per capta do Brasil')
plt.xlabel('Ano')
plt.ylabel('Valor do PIB')
```

```python
plt.figure(figsize=(12,4))
plt.scatter(df['lifeExp'], df['gdpPercap'], cmap = 'viridis')
plt.xlabel('Expectativa de Vida')
plt.ylabel('Renda per capta')
plt.title('Relação entre renda percapta e expectativa de vida', loc='left')
plt.show();
```

```python
plt.bar(x=df.index, height=df['pop'], color='magenta')
plt.xlabel("Anos")
plt.ylabel('População')
plt.title('População Brasileira')
```

```python
# Filtrando o continente
def filtrar_continente(continente):
    df = px.data.gapminder()
    df = df[df['continent'] == continente]
    return df
```

```python
# Função para filtrar o país
def filtrar_pais(pais, variavel):
    df = px.data.gapminder()
    df = df[df['country'] == pais][variavel]
    return df
```

```python
americas = filtrar_continente('Americas')
paises = americas['country'].unique()

# declara a figura
plt.figure(figsize=(12,8))

for pais in paises:
    plt.scatter(filtrar_pais(pais = pais, 
                             variavel = 'lifeExp'
                            ), 
               filtrar_pais (pais = pais, 
                             variavel='gdpPercap'
                            ))
plt.legend(labels = paises, loc = 'best')
plt.xlabel('Expectativa de vida')
plt.ylabel('Renda per capta')
plt.title('Relação entre renda per capta e expecativa de vida');
```
